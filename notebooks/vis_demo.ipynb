Notice that in both dataframes, the parent-child relationship are encoded by the 'msLevel' and 'MSnParentPeakID' columns, telling us the MSn level of each peak (1 or 2) and which child peaks are derived from which parent peaks.

In standard LDA, the input is the count of word occurences in documents. Similarly, we can turn our peak features data into something like documents-words by binning the child peaks by their masses. This turns continuous mass values into a discrete thing. For example, the child peaks with peakID 3 (mz 70.065056) and peakID 5 (mz 70.065071) are both binned into the word "70.06514". This 'word' can be found in the 'documents' corresponding to parent peaks with peakID = 1 and peakID = 4. 

There are two types of words (features) that we extracted from the data through this binning process:

* fragment words are the discretised values of mz of MS2 peak
* loss words are the discretised values of (mz of parent MS1 peak - mz of MS2 peak)

The input to LDA is the count of word occurences in documents (shown below). Each row is a document (encoded in mz_rt_peakID format of the originating MS1 peak) while each column is a word. Intensities of the MS2 peaks are taken as a proxy to the word counts, scaled from 0 .. 100.

```python
display(ms2lda.df)
```

Visualising the count matrix, we see that it's very sparse (full of zeroes). This is used as input to LDA.

```python
plt.figure()
plt.matshow(ms2lda.df)
plt.colorbar()
```

<h1>2. LDA Output</h1>

The main results from inference on the LDA model are **(1)** the document-topic matrix, which tells us the distribution of topics over **all documents** in the corpus and **(2)** the topic-word matrix, which tells us the distribution of topics over **all words** in the vocabulary. We can access them as below:

```python
doc_topic = ms2lda.model.doc_topic_
topic_word = ms2lda.model.topic_word_
```

doctopic is 1588 x 300, which is the no. of documents X no. of topics.

topicword is 300 x 3171, which is the no. of topics X no. of words.

```python
print doc_topic.shape
print topic_word.shape
```

```python
plt.figure()
plt.matshow(doc_topic)
plt.colorbar()
```

```python
plt.figure()
plt.matshow(topic_word)
plt.colorbar()
```

We can print the top words (having the highest probability) conditioned on the topic. Below we print for top 10.

```python
vocab = ms2lda.vocab
print vocab
```

```python
n_top_words = 10
for i, topic_dist in enumerate(topic_word):
    topic_words = np.array(vocab)[np.argsort(topic_dist)][:-n_top_words:-1]
    print('Topic {}: {}'.format(i, ' '.join(topic_words)))
```
